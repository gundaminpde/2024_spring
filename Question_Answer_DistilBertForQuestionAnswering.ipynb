```python
import transformers

print(transformers.__version__)
```

```python
from transformers import pipeline

QnA = pipeline("question-answering", model='distilbert-base-cased-distilled-squad')
##QnA = pipeline("question-answering")
```

```python
QnA.model
```

```python

context = "DistilBERT is a lighter version of BERT. It has fewer parameters and is designed for faster inference."
question = "What is DistilBERT?"
```

```python
answer = QnA(question=question, context=context)

print(answer)
```
